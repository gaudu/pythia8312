```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Files from Felix RIehn for hadronic interaction model dataset: Pythia 8.3.12 and Sibyll 2.3d.

```python
columns = ["event_number", "particle_stack_entry", "particle_name", "pdg_id", "pseudorapidity", "rapidity", "energy", "px", "py", "pz", "pt", "particle_charge"]

df_pythia = pd.read_csv("compare_c8_pp_output/particles-pythia8-p+-p+-10000GeV-lab.txt", delim_whitespace=True, header=None, names=columns, index_col=False)
df_sibyll = pd.read_csv("compare_c8_pp_output/particles-sibyll-p+-p+-10000GeV-lab.txt", delim_whitespace=True, header=None, names=columns, index_col=False)
```

### Particle multiplicity per particle type

```python
pdg_ids_pythia = df_pythia['pdg_id'].unique()
pdg_ids_sibyll = df_sibyll['pdg_id'].unique()
pythia_name_dict = df_pythia.groupby('pdg_id')['particle_name'].unique().to_dict()
sibyll_name_dict = df_sibyll.groupby('pdg_id')['particle_name'].unique().to_dict()
unique_pdg_ids = np.union1d(pdg_ids_pythia, pdg_ids_sibyll)

conflicts = []

for pdg_id in unique_pdg_ids:
    pythia_names = pythia_name_dict.get(pdg_id, [])
    sibyll_names = sibyll_name_dict.get(pdg_id, [])
    
    if pythia_names and sibyll_names and not np.array_equal(pythia_names, sibyll_names):
        conflicts.append((pdg_id, pythia_names, sibyll_names))
        
if conflicts:
    print("Conflicts found:")
    for conflict in conflicts:
        pdg_id, pythia_names, sibyll_names = conflict
        print(f"PDG ID {pdg_id} has different names:")
        print(f"  - Pythia: {pythia_names}")
        print(f"  - Sibyll: {sibyll_names}")
else:
    print("No conflicts found")
```

```python
def get_bin_edges(data):
    if len(data) == 0:
        return [0, 1]
    min_val = np.floor(data.min())
    max_val = np.ceil(data.max()) + 1
    if min_val >= max_val:
        min_val = max_val - 1
    return np.arange(min_val, max_val + 1, 1)
```

```python
unique_pdg_ids = np.union1d(df_pythia['pdg_id'].unique(), df_sibyll['pdg_id'].unique())
unique_pdg_ids = sorted(unique_pdg_ids, key=abs)
num_pdg_ids = len(unique_pdg_ids)
num_cols = 3
num_rows = (num_pdg_ids + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 3 * num_rows), constrained_layout=True)
axes = axes.flatten() if num_rows > 1 else [axes]

for i, pdg_id in enumerate(unique_pdg_ids):
    filtered_pythia = df_pythia[df_pythia['pdg_id'] == pdg_id]
    filtered_sibyll = df_sibyll[df_sibyll['pdg_id'] == pdg_id]
    particle_name_pythia = pythia_name_dict.get(pdg_id, ['Unknown'])[0]
    particle_name_sibyll = sibyll_name_dict.get(pdg_id, ['Unknown'])[0]
    particle_name = particle_name_pythia if particle_name_pythia != 'Unknown' else particle_name_sibyll
    particle_counts_pythia = filtered_pythia.groupby('event_number').size()
    particle_counts_sibyll = filtered_sibyll.groupby('event_number').size()
    bin_edges_pythia = get_bin_edges(particle_counts_pythia)
    bin_edges_sibyll = get_bin_edges(particle_counts_sibyll)
    axes[i].hist(particle_counts_pythia, bins=bin_edges_pythia, histtype='step', linewidth=2, color='tab:orange', label='Pythia 8.3.12')
    axes[i].hist(particle_counts_sibyll, bins=bin_edges_sibyll, histtype='step', linewidth=2, color='tab:blue', label='Sibyll 2.3d')
    
    axes[i].set_xlabel("Number of Particles per Event")
    axes[i].set_ylabel("")
    axes[i].set_title(f"{particle_name} (pdg id = {pdg_id})")
    axes[i].legend()
    
for j in range(num_pdg_ids, len(axes)):
    axes[j].axis('off')

fig.suptitle('Particle multiplicity distributions by particle type', fontsize=16)
```
