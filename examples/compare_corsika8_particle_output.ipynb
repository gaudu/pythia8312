### Rapidity

```python
plt.figure(figsize=(10, 6))
plt.hist(df_pythia["rapidity"], bins=100, histtype='step', linewidth=2, edgecolor='tab:orange', label='Pythia 8.3.12')
plt.hist(df_sibyll["rapidity"], bins=100, histtype='step', linewidth=2, edgecolor='tab:blue', label='Sibyll 2.3d')
plt.hist(df_epos["rapidity"], bins=100, histtype='step', linewidth=2, edgecolor='tab:red', label='EPOS-LHC')
plt.hist(df_qgsjet["rapidity"], bins=100, histtype='step', linewidth=2, edgecolor='tab:green', label='QGSJet-II.04')

plt.xlabel("Rapidity")
plt.ylabel("")
plt.xlim(-10, 20)
plt.grid(True)
plt.legend
```

EPOS-LHC predicts a peak at a higher rapidity (around 3), indicating more forward-boosted particles, possibly due to its inclusion of hydrodynamic effects and a different treatment of fragmentation and collective phenomena.

Pythia 8.3.12 predicts a peak at a lower rapidity (around 2), suggesting less forward particle production, which might be due to its focus on hard scattering and parton showers without as much emphasis on collective medium effects.

### Pseudorapidity

```python
plt.figure(figsize=(10, 6))
plt.hist(df_pythia["pseudorapidity"], bins=100, histtype='step', linewidth=2, edgecolor='tab:orange', label='Pythia 8.3.12')
plt.hist(df_sibyll["pseudorapidity"], bins=100, histtype='step', linewidth=2, edgecolor='tab:blue', label='Sibyll 2.3d')
plt.hist(df_epos["pseudorapidity"], bins=100, histtype='step', linewidth=2, edgecolor='tab:red', label='EPOS-LHC')
plt.hist(df_qgsjet["pseudorapidity"], bins=100, histtype='step', linewidth=2, edgecolor='tab:green', label='QGSJet-II.04')


plt.xlabel("Pseudorapidity")
plt.ylabel("")
plt.xlim(-15, 40)
plt.grid(True)
plt.legend()
```

The first peak around η=5 from EPOS-LHC: represents a significant number of secondary particles produced at forward rapidity, which is typical for secondary particles produced in high-energy collisions where the center of mass moves forward due to the high energy of the incoming particles.

The sharp peak at η≈25 from Pythia 8.3.12 is likely due to beam remnants or forward-scattered particles that retain much of the original beam momentum, thus moving nearly parallel to the beam axis.

### Particle multiplicity per particle type

```python
def get_bin_edges(data):
    if len(data) == 0:
        return [0, 1]
    min_val = np.floor(data.min())
    max_val = np.ceil(data.max()) + 1
    if min_val >= max_val:
        min_val = max_val - 1
    return np.arange(min_val, max_val + 1, 1)
```

```python
pdg_ids_pythia = df_pythia['pdg_id'].unique()
pdg_ids_sibyll = df_sibyll['pdg_id'].unique()
pdg_ids_epos = df_epos['pdg_id'].unique()
pdg_ids_qgsjet = df_qgsjet['pdg_id'].unique()
unique_pdg_ids = np.union1d(np.union1d(pdg_ids_pythia, pdg_ids_sibyll), np.union1d(pdg_ids_epos, pdg_ids_qgsjet))

pythia_name_dict = df_pythia.groupby('pdg_id')['particle_name'].unique().to_dict()
sibyll_name_dict = df_sibyll.groupby('pdg_id')['particle_name'].unique().to_dict()
epos_name_dict = df_epos.groupby('pdg_id')['particle_name'].unique().to_dict()
qgsjet_name_dict = df_qgsjet.groupby('pdg_id')['particle_name'].unique().to_dict()

conflicts = []

for pdg_id in unique_pdg_ids:
    pythia_names = pythia_name_dict.get(pdg_id, [])
    sibyll_names = sibyll_name_dict.get(pdg_id, [])
    epos_names = epos_name_dict.get(pdg_id, [])
    qgsjet_names = qgsjet_name_dict.get(pdg_id, [])
    
    all_names = np.unique(np.concatenate((pythia_names, sibyll_names, epos_names, qgsjet_names)))
    
    if len(all_names) > 1:
        conflicts.append((pdg_id, pythia_names, sibyll_names, epos_names, qgsjet_names))

if conflicts:
    print("Conflicts found:")
    for conflict in conflicts:
        pdg_id, pythia_names, sibyll_names, epos_names, qgsjet_names = conflict
        print(f"PDG ID {pdg_id} has different names:")
        print(f"  - Pythia: {pythia_names}")
        print(f"  - Sibyll: {sibyll_names}")
        print(f"  - EPOS: {epos_names}")
        print(f"  - QGSJET: {qgsjet_names}")
else:
    print("No conflicts found")
    
    if pythia_names and sibyll_names and not np.array_equal(pythia_names, sibyll_names):
        conflicts.append((pdg_id, pythia_names, sibyll_names))
```

```python
unique_pdg_ids = np.union1d(np.union1d(df_pythia['pdg_id'].unique(), df_sibyll['pdg_id'].unique()),
                            np.union1d(df_epos['pdg_id'].unique(), df_qgsjet['pdg_id'].unique()))
unique_pdg_ids = sorted(unique_pdg_ids, key=abs)
num_pdg_ids = len(unique_pdg_ids)

num_cols = 3
num_rows = (num_pdg_ids + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 3 * num_rows), constrained_layout=True)
axes = axes.flatten() if num_rows > 1 else [axes]

for i, pdg_id in enumerate(unique_pdg_ids):
    filtered_pythia = df_pythia[df_pythia['pdg_id'] == pdg_id]
    filtered_sibyll = df_sibyll[df_sibyll['pdg_id'] == pdg_id]
    filtered_epos = df_epos[df_epos['pdg_id'] == pdg_id]
    filtered_qgsjet = df_qgsjet[df_qgsjet['pdg_id'] == pdg_id]

    particle_name_pythia = pythia_name_dict.get(pdg_id, ['Unknown'])[0]
    particle_name_sibyll = sibyll_name_dict.get(pdg_id, ['Unknown'])[0]
    particle_name_epos = epos_name_dict.get(pdg_id, ['Unknown'])[0]
    particle_name_qgsjet = qgsjet_name_dict.get(pdg_id, ['Unknown'])[0]
    
    particle_name = particle_name_pythia
    if particle_name == 'Unknown':
        particle_name = particle_name_sibyll
    if particle_name == 'Unknown':
        particle_name = particle_name_epos
    if particle_name == 'Unknown':
        particle_name = particle_name_qgsjet

    particle_counts_pythia = filtered_pythia.groupby('event_number').size()
    particle_counts_sibyll = filtered_sibyll.groupby('event_number').size()
    particle_counts_epos = filtered_epos.groupby('event_number').size()
    particle_counts_qgsjet = filtered_qgsjet.groupby('event_number').size()

    bin_edges_pythia = get_bin_edges(particle_counts_pythia)
    bin_edges_sibyll = get_bin_edges(particle_counts_sibyll)
    bin_edges_epos = get_bin_edges(particle_counts_epos)
    bin_edges_qgsjet = get_bin_edges(particle_counts_qgsjet)

    axes[i].hist(particle_counts_pythia, bins=bin_edges_pythia, histtype='step', linewidth=2, color='tab:orange', label='Pythia 8.3.12')
    axes[i].hist(particle_counts_sibyll, bins=bin_edges_sibyll, histtype='step', linewidth=2, color='tab:blue', label='Sibyll 2.3d')
    axes[i].hist(particle_counts_epos, bins=bin_edges_epos, histtype='step', linewidth=2, color='tab:red', label='EPOS LHC')
    axes[i].hist(particle_counts_qgsjet, bins=bin_edges_qgsjet, histtype='step', linewidth=2, color='tab:green', label='QGSJet-II-04')

    axes[i].set_xlabel("Number of Particles per Event")
    axes[i].set_ylabel("")
    axes[i].set_title(f"{particle_name} (pdg id = {pdg_id})")
    axes[i].legend()

for j in range(num_pdg_ids, len(axes)):
    axes[j].axis('off')

fig.suptitle('Particle multiplicity distributions by particle type', fontsize=16)
```
